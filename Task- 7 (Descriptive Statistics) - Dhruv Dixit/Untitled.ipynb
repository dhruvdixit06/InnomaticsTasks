<h1 align='center'><u>Descriptive Statistics</u></h1>

<span align='center'>Descriptive statistics is a means of describing features of a data set by generating summaries about data samples. There are many methods for the same like mean, median, mode etc.</span>

<h1><span style="color:Red"><u>MEAN</u></span></h1>

<span>The mean of a dataset is the sum of all values divided by the total number of values. It's the most commonly used measure of central tendency and is often referred to as the “average.”</span>

<h3>Example:</h3>

<h3>Formulae:</h3>

<h2>Population Mean: $\mu$  $$ \mu = \sum\frac{x}{N} $$ </h2>
<br>
here, x= total of the numbers<br>
       N= total numbers

<h2>Sample Mean: $\bar x $ $$\bar x = \sum\frac{x}{n} $$ </h2>
<br>
here, x=total of the numbers<br> 
       n= total numbers

<h2>Geometric Mean: $$ GM(x) = \sqrt[n]{x_1\times x_2\times ...\times x_n} $$ </h2>

<h2>Harmonic Mean: $$ HM(x) = \frac{n}{\frac{1}{x_1}+\frac{1}{x_2}+...+\frac{1}{x_n}}$$ </h2>

<h1><span style="color:Red"><u>MODE</u></span></h1>

<span>A mode is defined as the value that has a higher frequency in a given set of values. It is the value that appears the most number of times.</span>

<h3>Example:</h3>

<h3>Formulae:</h3>

<h2> $$ mode = l+h\left(\frac{f_m - f_1}{2f_m - f_1 - f_2}\right) $$</h2>
<br>
here,<br>
l=the lower limit of the modal class<br>
h= the size of the class interval<br>
$f_m$=the frequency of the modal class<br>
$f_1$=the frequency of the class preceding the modal class<br>
$f_2$=the frequency of the class succeeding the modal class

<h3>Code Examples:</h3><br><u>With Library Function:</u>

```python
df.mode()
```

```python
MthlyHH = list(df['Mthly_HH_Income'])
MthlyHH 
```

```python
from collections import Counter
```

```python
count = Counter(MthlyHH )
count
```

```python
def mode(counter):
    # store a list of name:count tuples in case multiple modes
    modes = [('',0)]
    for k,v in counter.items():
        highest_count = modes[0][1]
        if v > highest_count:
            modes = [(k,v)]
        elif v == highest_count:
            modes.append((k,v))
        
    return modes
mode(count)
```

<h1><span style="color:Red"><u>VARIANCE</u></span></h1>

<span>The term variance refers to a statistical measurement of the spread between numbers in a data set. More specifically, variance measures how far each number in the set is from the mean (average), and thus from every other number in the set.</span>

<h3>Example:</h3>

<h3>Formulae:</h3>

<h2>Population Variance: $$ \sigma^2 = \frac{\sum\,(X_i - \mu)^2}{N} $$ </h2>
<br>
here,<br>
$\sigma^2$ = population variance<br>
$\sum$ = sum <br>
Χ = each value<br>
$\mu$ = population mean<br>
Ν = number of values in the population<br>

<h2>Sample Variance : $$ \sigma^2 = \frac{\sum\,(X_i - \bar{x})^2}{n-1} $$ </h2>
<br>
here,<br>
$\sigma^2$ = sample variance<br>
$\sum$ = sum <br>
Χ = each value<br>
$\bar{x}$ = sample mean<br>
n = number of values in the sample

<h3>Code Examples:</h3><br><u>With Library Function:</u>

<h1><span style="color:Red"><u>STANDARD DEVIATION</u></span></h1>

<span>Standard deviation is a statistic that measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance.</span>

<h3>Example:</h3>

<h3>Formulae:</h3>

<h2>Population Standard Deviation: $$ \sigma = \sqrt{\frac{\sum\,(X_i - \mu)^2}{N}} $$ </h2>
<br>
here,<br>
$\sigma$=Standard Deviation<br>
X=Terms given in the data<br>
N=Total number of terms<br>
$\mu$=Mean of the population

<h2>Sample Standard Deviation : $$  \sigma = \sqrt{\frac{\sum\,(X_i - \bar{x})^2}{N-1}} $$ </h2>
<br>
here,<br>
$\sigma$=Standard Deviation<br>
X=Terms given in the data<br>
N=Total number of terms<br>
$\bar x $= Mean of the data

<h3>Code Examples:</h3><br><u>With Library Function:</u>

<h1><span style="color:Red"><u>CORRELATION</u></span></h1>

<span>Correlation is a statistical measure (expressed as a number) that describes the size and direction of a relationship between two or more variables. A correlation between variables, however, does not automatically mean that the change in one variable is the cause of the change in the values of the other variable.</span>

<h3>Example:</h3>

<h2><u>Different types of Correlation:</u></h2>

<h3><u>Pearson Correlation</u></h3>

<span>The Pearson correlation measures the strength of the linear relationship between two variables. It has a value between -1 to 1, with a value of -1 meaning a total negative linear correlation, 0 being no correlation, and + 1 meaning a total positive correlation.</span>

<h3>Formula:</h3>

<h2>Pearson Correlation Coefficient:   $$ r = \frac{\sum(x_i-\bar x)(y_i-\bar y)}{\sqrt{\sum(x_i-\bar x)^2\sum(y_i-\bar y)^2}} $$ </h2>
<br>
here,<br>
$x_i$=x variable sample<br>
$y_i$=y variable sample<br>
$\bar x$=mean of values in x variable<br>
$\bar y$=mean of values in y variable<br>

<h3>Example</h3>

```python
from scipy.stats import pearsonr
```

```python
a = np.random.normal(size=10)
b = np.random.normal(size=10)
```

```python
pearsonr(a,b)
```

<h3><u>Spearman’s Rank Correlation</u></h3>

<span>Spearman's rank correlation measures the strength and direction of association between two ranked variables. It basically gives the measure of monotonicity of the relation between two variables i.e. how well the relationship between two variables could be represented using a monotonic function.</span>

<h3>Formula:</h3>

<h2>Spearman’s Rank Correlation Coefficient: $$\rho = 1- \frac{6\sum d_i ^2}{n(n^2 -1)}$$</h2>
<br>
here,<br>
$d_i$=difference between the two ranks of each observation<br>
n= number of observations

<h3>Example:</h3>

```python
from scipy.stats import spearmanr
```

<h1><span style="color:Red"><u>NORMAL DISTRIBUTION</u></span></h1>

<span>Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graphical form, the normal distribution appears as a "bell curve".</span>

<h3>Example:</h3>

<h3>Formula:</h3>

<h2>Normal Distribution:</h2>
<h2>$$ f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^\left(-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{2}\right) $$ </h2>
<br>
here,<br>
μ = mean
σ = standard deviation
x = value of the variable or data being examined<br>
f(x)= the probability function
<p><i>the values of the mean (μ) and standard deviation(σ)are necessary</i></p>

<h3>Code Example:</h3>

```python
from scipy.stats import norm
cdf_upper_limit = norm(loc = 5.3 , scale = 1).cdf(6.5)
cdf_lower_limit = norm(loc = 5.3 , scale = 1).cdf(4.5)
 
prob = cdf_upper_limit - cdf_lower_limit
```

<h1><span style="color:Red"><u>FEATURE OF NORMAL DISTRIBUTION</u></span></h1>

<span style="color:blue">1. It is symmetric:</span><br>
A normal distribution comes with a perfectly symmetrical shape. This means that the distribution curve can be divided in the middle to produce two equal halves. The symmetric shape occurs when one-half of the observations fall on each side of the curve.
<br><br>
<span style="color:blue">2. The mean, median, and mode are equal:</span><br>
The middle point of a normal distribution is the point with the maximum frequency, which means that it possesses the most observations of the variable. The midpoint is also the point where these three measures fall. The measures are usually equal in a perfectly (normal) distribution.
<br><br>
<span style="color:blue">3. Empirical rule:</span><br>
In normally distributed data, there is a constant proportion of distance lying under the curve between the mean and specific number of standard deviations from the mean. For example, 68.25% of all cases fall within +/- one standard deviation from the mean. 95% of all cases fall within +/- two standard deviations from the mean, while 99% of all cases fall within +/- three standard deviations from the mean.
<br><br>
<span style="color:blue">4. Skewness and kurtosis:</span><br>
Skewness and kurtosis are coefficients that measure how different a distribution is from a normal distribution. Skewness measures the symmetry of a normal distribution while kurtosis measures the thickness of the tail ends relative to the tails of a normal distribution.<br>

<h1><span style="color:Red"><u>POSITIVELY SKEWED AND NEGATIVELY SKEWED NORMAL DISTRIBUTION</u></span></h1>

<h3>Left Skewed:</h3><p>A distribution is left skewed if it has a “tail” on the left side of the distribution. In this, Mean &lt Median &lt Mode.</p>

<h3>Right Skewed:</h3><p>A distribution is right skewed if it has a “tail” on the right side of the distribution. In this, Mode &lt Median &lt Mean.</p>

<h3>Right Skewed:</h3><p>A distribution has no skew if it’s symmetrical on both sides. In this, Mean = Median = Mode.</p>

<h3>Formulae:</h3>

<h2>Pearson's first skewness coefficient (mode skewness):<br><br>$$ \frac{Mean - Mode}{StandardDeviation}$$</h2>

<h2>Pearson's second skewness coefficient (median skewness):<br><br> $$ \frac{3(Mean - Median)}{StandardDeviation}$$

<h1><span style="color:Red"><u>Box Cox Transformation</u></span></h1>

<span>A Box Cox transformation is a transformation of non-normal dependent variables into a normal shape. Normality is an important assumption for many statistical techniques; if your data isn't normal, applying a Box-Cox means that you are able to run a broader number of tests.</span>

<h3>Formula:</h3>

<h2>Transformation Equation:<br><br>
$$ y(\lambda) = \left\{\begin{array}\\ \left(y^ \lambda -1\right)/\lambda & \mbox{if}\ \lambda \neq {0}\\
\log {y} & \mbox{if}\ \lambda = 0 \end{array}\right. $$ </h2>

<h3>Code Examples:</h3>

```python
plt.figure(figsize = (8, 8))
data = df['Mthly_HH_Expense'] 
sb.histplot(data,kde=True)
plt.show()
```

```python
from scipy import stats
tdata = stats.boxcox(df['Mthly_HH_Expense'])
plt.figure(figsize = (8, 8))
sb.histplot(tdata,kde=True)
plt.show()
```
